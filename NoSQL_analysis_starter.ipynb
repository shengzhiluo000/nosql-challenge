### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query1 = {'scores.Hygiene': 20}
#fields1 = {'BusinessName':1, 'scores.Hygiene':1}
result1 = establishments.find(query1)

# Use count_documents to display the number of documents in the result
print("Number of documents in result:", establishments.count_documents(query1))

# Display the first document in the results using pprint
pprint(result1[0])
```

```python
# Convert the result to a Pandas DataFrame
df1 = pd.DataFrame(result1)
# Display the number of rows in the DataFrame
print(df1.count())
# Display the first 10 rows of the DataFrame
df1.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query2 = {'LocalAuthorityName':{'$regex':'London'},'RatingValue':{'$gte':4}}
fields2 = {'BusinessName':1, 'LocalAuthorityName':1, 'RatingValue':1}
result2 = establishments.find(query2,fields2)

# Use count_documents to display the number of documents in the result
print("Number of documents in result:", establishments.count_documents(query2))

# Display the first document in the results using pprint
pprint(result2[0])
```

```python
# Convert the result to a Pandas DataFrame
df2 = pd.DataFrame(result2)
# Display the number of rows in the DataFrame
print(df2.count())
# Display the first 10 rows of the DataFrame
df2.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
establishments.find({'BusinessName':'Penang Flavours'})[0]['geocode']['longitude']
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = establishments.find({'BusinessName':'Penang Flavours'})[0]['geocode']['latitude']
longitude = establishments.find({'BusinessName':'Penang Flavours'})[0]['geocode']['longitude']
#,'$lte':longitude+degree_search}
query = {'RatingValue':5,
         'geocode.latitude':{'$gte':latitude-degree_search, '$lte':latitude+degree_search},
         'geocode.longitude':{'$gte':longitude-degree_search,'$lte':longitude+degree_search}}

sort =  [('scores.Hygiene',1)]

result3 = list(establishments.find(query).sort(sort))
# Print the results
print("Number of documents in result:", establishments.count_documents(query))
pprint(result3)
```

```python
# Convert result to Pandas DataFrame
df3=pd.DataFrame(result3)
df3
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match_query = {'$match':{'scores.Hygiene':0}}

group_query = {'$group':{'_id':'$LocalAuthorityName','count': { '$sum': 1 }}}

sort_query = {'$sort': {'count': -1}}

pipeline = [match_query, group_query, sort_query]

result4 = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in result:",len(result4))

# Print the first 10 results
pprint(result4)
```

```python
# Convert the result to a Pandas DataFrame
df4=pd.DataFrame(result4)

# Display the number of rows in the DataFrame
print(df4.count())

# Display the first 10 rows of the DataFrame
df4.head(10)
```
